```python
# 기초분석(기술 통계량)

# count         NA를 제외한 갯수
# min, max      최소, 최대값
# sum           합
# comprod       누적합
# mean          평균
# median        중앙값
# quantile      분위수
# var           표본분산
# std           표본 정규분산
# describe      요약 통계량
```

```python
german_sample = pd.read_csv('http://freakonometrics.free.fr/german_credit.csv')
type(german_sample)
```

```python
list(german_sample.columns.values)
```

```python
german = german_sample[['Creditability','Duration of Credit (month)',
               'Purpose','Credit Amount']]
german
```

```python
german.min()
```

```python
german.max
```

```python
german.mean()
```

```python
german.describe
```

```python
#상관관계
german = german_sample[['Duration of Credit (month)',
                       'Credit Amount',
                       'Age (years)']]
german.head()
```

```python
german.corr()
```

```python
#핵심기능 group by
german = german_sample[['Credit Amount','Type of apartment']]
german.head()
```

```python
german_grouped = german['Credit Amount'].groupby(german['Type of apartment'])
```

```python
german_grouped.mean()
```

```python
german = german_sample[['Credit Amount','Type of apartment','Purpose']]
german_group2=german['Credit Amount'].groupby([german['Purpose'],german['Type of apartment']])

german_group2.mean()
```

```python
german = german_sample[['Type of apartment','Sex & Marital Status', 'Credit Amount']]
for g in german.groupby('Type of apartment'):

  print(g)
  print('--------------------------------------------------------------------------------')
```

```python
births=pd.read_csv('https://raw.githubusercontent.com/jakevdp/data-CDCbirths/master/births.csv')
births.head()
```

```python
births['year']/10
```

```python
births['decade'] = births['year'] // 10*10
births.head
```

```python
births.pivot_table('births', index='decade', columns='gender',aggfunc='sum')
```

```python
births.pivot_table('births', index='decade', columns='gender',aggfunc='sum').plot()
```

```python
births.pivot_table('births', index='year', columns='gender',aggfunc='sum').plot()
```
