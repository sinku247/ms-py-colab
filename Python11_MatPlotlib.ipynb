```python
plt.subplot(2,1,1)    #rows, columns, panel number
plt.plot(x,np.sin(x))
plt.subplot(2,1,2)
plt.plot(x,np.cos(x))
```

```python
plt.plot(x, np.sin(x),color='blue')
plt.plot(x, np.sin(x-1),color='g')
plt.plot(x, np.sin(x-2),color='0.75')
plt.plot(x, np.sin(x-3),color='#FFDD44')
plt.plot(x, np.sin(x-4),color=(1.0,0.2,0.3))
plt.plot(x, np.sin(x-5),color='chartreuse')
```

```python
plt.plot(x, x+0, linestyle='solid')
plt.plot(x, x+1, linestyle='dashed')
plt.plot(x, x+2, linestyle='dashdot')
plt.plot(x, x+3, linestyle='dotted')

plt.plot(x, x+4, linestyle='-')
plt.plot(x, x+5, linestyle='--')
plt.plot(x, x+6, linestyle='-.')
plt.plot(x, x+7, linestyle=':')
```

```python
plt.plot(x, x+0, '-g')            #solid green
plt.plot(x, x+1, '--c')           #dashed cyan
plt.plot(x, x+2, '-.k')           #dashdot black
plt.plot(x, x+3, '-.r')           #dashdot red
```

```python
plt.plot(x,np.sin(x))       #리미트 설정이 없으면 최대한 꽉 채워서 보여주려함.
plt.xlim(-1,11)
plt.ylim(-1.5,1.5)
```

```python
plt.plot(x, np.sin(x))        #상하반전, 좌우반전전
plt.xlim(10,0)
plt.ylim(1.2,-1.2)
```

```python
plt.plot(x,np.sin(x))
plt.axis([-1,11,-1.5,1.5])
```

```python
plt.plot(x, np.sin(x))
plt.axis('tight')     #타이트하게 출력
```

```python
plt.plot(x,np.sin(x))
plt.axis('equal')         #x, y 비율을 동일하게 준다.
```

```python
#label 붙이기
plt.plot(x, np.sin(x))
plt.title('A Sine Curve')
plt.xlabel('x')
plt.ylabel('sin(x)')
```

```python
plt.plot(x, np.sin(x),'-g', label='sin(x)')
plt.plot(x, np.cos(x),':b', label='cos(x)')
plt.axis('equal')
plt.legend()
plt.xlabel('x')
plt.ylabel('y')
plt.title('Sin and Cos')
```

```python
plt.style.use('bmh')
rng = np.random.RandomState(0)
for marker in ['o',',',',','x','+','v','^','<','>','s','d']:
  plt.plot(rng.rand(5), rng.rand(5), marker, label='marker={0}'.format(marker))
  plt.legend(numpoints=1)
  plt.xlim(0, 1.8)
```

```python
plt.style.use('classic')
rng=np.random.RandomState(0)
x=rng.randn(100)              
y=rng.randn(100)
color=rng.rand(100)
sizes=rng.rand(100) * 1000
plt.scatter(x,y,s=sizes,c=color,alpha=0.3,cmap='viridis')
```

```python
from sklearn.datasets import load_iris
iris=load_iris()
features=iris.data.T

plt.scatter(features[2],features[3],c=iris.target,s=features[3]*100, alpha=0.2, cmap='viridis')
plt.xlabel(iris.feature_names[2])
plt.ylabel(iris.feature_names[3])
iris.feature_names[3]
```

```python
# 오차 시각화 하기
x=np.linspace(0,10,50)
dy = 0.8
y=np.sin(x+np.random.randn(50))

plt.errorbar(x,y,yerr=dy, fmt='.k')
```

```python
plt.errorbar(x,y, yerr=dy,fmt='o',color='black')
```

```python
plt.errorbar(x,y,yerr=dy,fmt='o',color='black',ecolor='lightgrey',elinewidth=3)
```
